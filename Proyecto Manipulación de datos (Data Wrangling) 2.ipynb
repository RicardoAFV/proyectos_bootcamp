- `order_products.csv`: cada fila corresponde a un artículo pedido en un pedido.
    - `'order_id'`: número de ID que identifica de manera única cada pedido.
    - `'product_id'`: número ID que identifica de manera única cada producto.
    - `'add_to_cart_order'`: el orden secuencial en el que se añadió cada artículo en el carrito.

- `aisles.csv`
    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.
    - `'aisle'`: nombre del pasillo.

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
total_orders = order_products["product_id"].value_counts()
reordered_counts = order_products[order_products["reordered"] == 1]["product_id"].value_counts()
```

```python
reorder_ratio = pd.DataFrame({
    "product_id": total_orders.index,
    "total_orders": total_orders.values,
    "reordered": reordered_counts.reindex(total_orders.index, fill_value=0).values
})
```

```python
reorder_ratio["reorder_proportion"] = reorder_ratio["reordered"] / reorder_ratio["total_orders"]
reorder_ratio = reorder_ratio.merge(products, on="product_id", how="left")
print(reorder_ratio[["product_id", "product_name", "total_orders", "reordered", "reorder_proportion"]].head(20))
```

### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

```python
orders_products = order_products.merge(orders[["order_id", "user_id"]], on="order_id", how="left")
total_products_per_user = orders_products.groupby("user_id")["product_id"].count()
reordered_products_per_user = orders_products[orders_products["reordered"] == 1].groupby("user_id")["product_id"].count()
```

```python
user_reorder_ratio = pd.DataFrame({
    "total_products": total_products_per_user,
    "reordered_products": reordered_products_per_user.reindex(total_products_per_user.index, fill_value=0)
})
user_reorder_ratio["reorder_proportion"] = user_reorder_ratio["reordered_products"] / user_reorder_ratio["total_products"]
print(user_reorder_ratio.head(20))
```

Escribe aquí tus conclusiones

Clientes con alta proporción de reordenamiento
Son compradores recurrentes que piden los mismos productos en la mayoría de sus pedidos.
Posiblemente compran productos de consumo frecuente como leche, pan y huevos.

Clientes con baja proporción de reordenamiento
Son clientes nuevos o que experimentan con productos distintos en cada compra.
También pueden ser clientes ocasionales que no tienen un patrón de compra definido.
