```python
# Converti nombres de columnas a minúsculas
df_ice_game_store.columns = df_ice_game_store.columns.str.lower()

# Elimine filas con valores nulos en 'name' y 'genre'
df_ice_game_store.dropna(subset=['name', 'genre'], inplace=True)

# Converti tipos de datos y maneje valores ausentes
df_ice_game_store['year_of_release'].fillna(df_ice_game_store['year_of_release'].median(), inplace=True)
df_ice_game_store['year_of_release'] = df_ice_game_store['year_of_release'].astype(int)
df_ice_game_store['critic_score'].fillna(df_ice_game_store['critic_score'].median(), inplace=True)
df_ice_game_store['user_score'] = pd.to_numeric(df_ice_game_store['user_score'], errors='coerce')
df_ice_game_store['user_score'].fillna('Sin puntuación', inplace=True)
df_ice_game_store['rating'].fillna('unknown', inplace=True)

# Calcule ventas totales
df_ice_game_store['total_sales'] = df_ice_game_store[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)

# Elimine duplicados
df_ice_game_store.drop_duplicates(inplace=True)

# Verifique de valores ausentes
print(df_ice_game_store.isna().sum())

print(df_ice_game_store)
```

```python
#Sumamos las ventas (total_sales) por plataforma para identificar las más exitosas:
platform_sales = df_ice_game_store.groupby('platform')['total_sales'].sum().reset_index()
platform_sales = platform_sales.sort_values(by='total_sales', ascending=False)

print(platform_sales)  
```

```python
#Analizamos las 10 plataformas con mas ventas y filtramos los datos para esas plataformas y agrupamos por año y plataforma.
top_platforms = platform_sales.head(10)['platform'].tolist() 
df_top_platforms = df_ice_game_store[df_ice_game_store['platform'].isin(top_platforms)]
sales_by_year = df_top_platforms.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()

print(top_platforms)
```

```python
#Hacemos un gráfico de líneas para ver la evolución de ventas de cada plataforma a lo largo del tiempo.
plt.figure(figsize=(12,6))
sns.lineplot(data=sales_by_year, x='year_of_release', y='total_sales', hue='platform', marker='o')

plt.title('Ventas Totales por Año para las Principales Plataformas')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Ventas Totales (millones)')
plt.legend(title='Plataforma')
plt.grid(True)
plt.show()
```

```python
#Verificamos cuál es el último año con registros en el dataset.
ultimo_año = df_ice_game_store['year_of_release'].max()
print(f"Último año con datos: {ultimo_año}")
```

```python
#Buscamos las plataformas que tuvieron ventas significativas en algún momento.
ventas_por_plataforma = df_ice_game_store.groupby('platform')['total_sales'].sum().reset_index()
ventas_por_plataforma = ventas_por_plataforma.sort_values(by='total_sales', ascending=False)

print(ventas_por_plataforma)
```

```python
# Encontramos las plataformas que tienen ventas en algún año anterior
plataformas_antiguas = df_ice_game_store[df_ice_game_store['year_of_release'] < ultimo_año]['platform'].unique()

# Encontramos plataformas que tienen ventas en el último año registrado
plataformas_actuales = df_ice_game_store[df_ice_game_store['year_of_release'] == ultimo_año]['platform'].unique()

# Filtramos plataformas que solían vender pero ahora no tienen ventas
plataformas_descontinuadas = set(plataformas_antiguas) - set(plataformas_actuales)

print(plataformas_descontinuadas)
```

```python
# Filtramos datos solo de las plataformas descontinuadas
df_descontinuadas = df_ice_game_store[df_ice_game_store['platform'].isin(plataformas_descontinuadas)]

# Agrupamos ventas por año y plataforma
ventas_descontinuadas = df_descontinuadas.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()

# Graficamos la evolución de ventas
plt.figure(figsize=(12,6))
sns.lineplot(data=ventas_descontinuadas, x='year_of_release', y='total_sales', hue='platform', marker='o')

plt.title('Declive de Ventas de Plataformas Descontinuadas')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Ventas Totales (millones)')
plt.legend(title='Plataforma')
plt.grid(True)
plt.show()
```

```python
# Encontramos el primer año de lanzamiento para cada plataforma
primer_año_plataforma = df_ice_game_store.groupby('platform')['year_of_release'].min().reset_index()

# Ordenamos por año de aparición
primer_año_plataforma = primer_año_plataforma.sort_values(by='year_of_release')

# Visualizamos los primeros años de aparición de plataformas
print(primer_año_plataforma)
```

```python
# Agrupe las ventas totales por plataforma y año
ventas_por_plataforma_y_año = df_ice_game_store.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()

# Defini un umbral para considerar una plataforma como "desaparecida" (sin ventas)
umbral_ventas = 0.1  # En millones, por ejemplo

# Filtre las plataformas que ya no tienen ventas
plataformas_desaparecidas = ventas_por_plataforma_y_año[ventas_por_plataforma_y_año['total_sales'] < umbral_ventas]['platform'].unique()

# Encontre el último año con ventas significativas para estas plataformas
ultimo_año_ventas = df_ice_game_store[df_ice_game_store['platform'].isin(plataformas_desaparecidas)].groupby('platform')['year_of_release'].max().reset_index()

# Calcule la duración de vida de cada plataforma
primer_año_plataforma = primer_año_plataforma[primer_año_plataforma['platform'].isin(plataformas_desaparecidas)]
duracion_vida_plataforma = pd.merge(primer_año_plataforma, ultimo_año_ventas, on='platform')

# Calcule la duración promedio de vida de las plataformas
duracion_vida_plataforma['duracion'] = duracion_vida_plataforma['year_of_release_y'] - duracion_vida_plataforma['year_of_release_x']
duracion_promedio = duracion_vida_plataforma['duracion'].mean()

print(f"La duración promedio de vida de las plataformas es: {duracion_promedio} años.")
```

```python
#Cree unos graficos donde muestan las ditribuciones de apariciones y duracion de vida de las platafomas
plt.figure(figsize=(12, 6))
sns.histplot(primer_año_plataforma['year_of_release'], bins=30, kde=True)
plt.title('Distribución del Año de Aparición de Nuevas Plataformas')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Plataformas')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(duracion_vida_plataforma['duracion'], bins=20, kde=True)
plt.title('Distribución de la Duración de Vida de las Plataformas')
plt.xlabel('Duración en Años')
plt.ylabel('Número de Plataformas')
plt.show()
```

```python
# Encontre el primer y último año de ventas por plataforma
plataforma_ciclo_vida = df_ice_game_store.groupby('platform')['year_of_release'].agg(['min', 'max']).reset_index()

# Renombre columnas para mayor claridad
plataforma_ciclo_vida.columns = ['platform', 'year_first_sale', 'year_last_sale']

print(plataforma_ciclo_vida)  
```

```python
#Analice la duracion del ciclo de vida en ventas de cada plataforma y se ordeno de la que mas años tubo a la que menos duro
plataforma_ciclo_vida['duration'] = plataforma_ciclo_vida['year_last_sale'] - plataforma_ciclo_vida['year_first_sale']

print(plataforma_ciclo_vida.sort_values(by='duration', ascending=False))  
```

```python
# Tiempo entre nuevas plataformas
tiempo_entre_nuevas = plataforma_ciclo_vida['year_first_sale'].sort_values().diff().dropna().mean()
print(f"Tiempo promedio entre la aparición de nuevas plataformas: {tiempo_entre_nuevas:.1f} años")

# Duración promedio de una plataforma antes de desaparecer
duracion_promedio = plataforma_ciclo_vida['duration'].mean()
print(f"Duración promedio de una plataforma antes de desaparecer: {duracion_promedio:.1f} años")
```

```python
plt.figure(figsize=(12,6))

for _, row in plataforma_ciclo_vida.iterrows():
    plt.plot([row['year_first_sale'], row['year_last_sale']], [row['platform'], row['platform']], marker='o')

plt.title('Ciclo de Vida de las Plataformas de Videojuegos')
plt.xlabel('Año')
plt.ylabel('Plataforma')
plt.grid(True)
plt.show()
```

En los datos podemos ver que la aparición de nuevas consolas afecta significativamente las ventas de las anteriores, pero estas aún se mantienen unos años.
El ciclo de vida de 9 años se confirma en la mayoría de los casos (PS2, PS3, Xbox 360, etc.).
Los mayores picos de ventas ocurrieron en 1994-2011, coincidiendo con el lanzamiento de las plataformas más populares.

```python
df_filtered_1994_2011 = df_ice_game_store[(df_ice_game_store['year_of_release'] >= 1994) & (df_ice_game_store['year_of_release'] <= 2011)]
platform_sales = df_filtered_1994_2011.groupby('platform')['total_sales'].sum()
platform_sales_sorted = platform_sales.sort_values(ascending=False)
print(platform_sales_sorted)
```

```python
df_1994_2011_no_outliers = df_filtered_1994_2011.copy()

# Crear una lista para guardar los índices sin outliers
indices_sin_outliers = []

# Agrupar por plataforma
for platform, grupo in df_1994_2011_no_outliers.groupby('platform'):
    q1 = grupo['total_sales'].quantile(0.25)
    q3 = grupo['total_sales'].quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr

    # Filtrar los que no son outliers
    sin_outliers = grupo[(grupo['total_sales'] >= limite_inferior) & (grupo['total_sales'] <= limite_superior)]
    indices_sin_outliers.extend(sin_outliers.index)

# Filtrar el DataFrame para quedarnos solo con los datos sin outliers
df_1994_2011_no_outliers = df_1994_2011_no_outliers.loc[indices_sin_outliers]

plt.figure(figsize=(15, 6))
sns.boxplot(data=df_1994_2011_no_outliers, x='platform', y='total_sales')
plt.title('Boxplot de ventas globales por plataforma (sin outliers)')
plt.xticks(rotation=90)
plt.ylabel('Ventas Globales (millones)')
plt.xlabel('Plataforma')
plt.tight_layout()
plt.show()
```

Paso 5. Prueba las siguientes hipótesis:

```python
df_ice_game_store["user_score"] = pd.to_numeric(df_ice_game_store["user_score"], errors="coerce")


xbox_scores = df_ice_game_store[df_ice_game_store["platform"] == "XOne"]["user_score"].dropna()
pc_scores = df_ice_game_store[df_ice_game_store["platform"] == "PC"]["user_score"].dropna()


stat_levene, p_levene = stats.levene(xbox_scores, pc_scores)
print(f"Prueba de Levene:\nEstadístico = {stat_levene:.4f}, Valor p = {p_levene:.4f}")


alpha = 0.05
if p_levene < alpha:
    print("Rechazamos H₀: Las varianzas son diferentes. Usamos equal_var=False.")
    equal_var_setting = False
else:
    print("No se rechaza H₀: Las varianzas son iguales. Usamos equal_var=True.")
    equal_var_setting = True


t_stat, p_value = stats.ttest_ind(xbox_scores, pc_scores, equal_var=equal_var_setting)
print(f"\nPrueba t (Xbox One vs PC):\nEstadístico t = {t_stat:.4f}, Valor p = {p_value:.4f}")

if p_value < alpha:
    print("Rechazamos H₀: Hay diferencia significativa en las calificaciones entre Xbox One y PC.")
else:
    print("No se rechaza H₀: No hay evidencia suficiente para afirmar que las calificaciones son diferentes.")
```

```python

action_scores = df_ice_game_store[df_ice_game_store["genre"] == "Action"]["user_score"].dropna()
sports_scores = df_ice_game_store[df_ice_game_store["genre"] == "Sports"]["user_score"].dropna()


stat_levene, p_levene = stats.levene(action_scores, sports_scores)
print(f"Prueba de Levene:\nEstadístico = {stat_levene:.4f}, Valor p = {p_levene:.4f}")


alpha = 0.05
if p_levene < alpha:
    print("Rechazamos H₀: Las varianzas son diferentes. Usamos equal_var=False.")
    equal_var_setting = False
else:
    print("No se rechaza H₀: Las varianzas son iguales. Usamos equal_var=True.")
    equal_var_setting = True


t_stat_action_sports, p_value_action_sports = stats.ttest_ind(
    action_scores, sports_scores, equal_var=equal_var_setting
)

print(f"\nPrueba t (Acción vs Deportes):\nEstadístico t = {t_stat_action_sports:.4f}, Valor p = {p_value_action_sports:.4f}")

if p_value_action_sports < alpha:
    print("Rechazamos H₀: Hay diferencia significativa en las calificaciones entre Acción y Deportes.")
else:
    print("No se rechaza H₀: No hay evidencia suficiente para afirmar que las calificaciones son diferentes.")
```
